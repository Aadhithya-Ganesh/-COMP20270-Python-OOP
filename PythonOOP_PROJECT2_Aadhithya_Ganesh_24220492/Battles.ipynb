This is the fight method to make a 2 list of creatures fight each other.
The game will end if either team is defeated by the other.

The Allies get the First turn.
The Enemies go next.

```python
def fight(allies, enemies):
    round = 1
    while round <= 20:
        print("Round", round, ":")
        for creature in allies:
            result = creature.turn(round, enemies)
            if type(result) != type(None) and type(result) != type("str"):
                enemies = list(filter(lambda item : item not in result, enemies))
            elif result == "No enemies":
                break
            
        for creature in enemies:
            result = creature.turn(round, allies)
            if type(result) != type(None) and type(result) != type("str"):
                allies = list(filter(lambda item : item not in result, allies))
            elif result == "No enemies":
                break
        
        if len(allies) == 0 or len(enemies) == 0:
            break

        print()
        round+=1

    print()
    print(allies)
    print(enemies)
```
